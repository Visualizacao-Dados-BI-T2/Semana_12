```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Exercício 1 - Gráfico de linha

Use: dataset_voos = sns.load_dataset("flights"). 

Pergunta: Como o número de passageiros mudou ao longo dos anos?  
Objetivo: Trabalhar tendência no tempo.


```python
dataset_voos = sns.load_dataset('flights')
```

```python
dataset_voos.head()
```

```python
sns.set_theme(style="darkgrid")

sns.lineplot(
    data=dataset_voos,
    x="year",
    y="passengers"
)

plt.title("Número de passageiros por ano")
plt.xlabel("Ano")
plt.ylabel("Número de passageiros")
plt.annotate('Ano de menor número de passageiros', xy=(1949, 130), xytext=(1954, 150),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Ano de maior número de passageiros', xy=(1960, 500), xytext=(1955, 300),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset
penguins = sns.load_dataset("penguins")

# Remover valores nulos para evitar problemas no gráfico
penguins = penguins.dropna()

# Criar o FacetGrid- Cria uma grade de gráficos, separando os dados pela coluna species
grid_species = sns.FacetGrid(
                        penguins,
                        row="species"
                            )

# Aplicar o histograma em cada gráfico - Aplica um histograma em cada gráfico, usando a variável bill_length_mm
grid_species.map(
    sns.histplot,
    "bill_length_mm"
)

# Ajustar títulos
grid_species.set_titles("{row_name}")

# Ajustar nomes dos eixos
grid_species.set_axis_labels(
    "Comprimento do bico (mm)",
    "Quantidade de pinguins"
)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset
penguins = sns.load_dataset("penguins")

# Remover valores nulos para evitar problemas no gráfico
penguins = penguins.dropna()

sns.set_theme(style="darkgrid")
# Criar o FacetGrid- Cria uma grade de gráficos, separando os dados pela coluna species
grid_species = sns.FacetGrid(
                        penguins,
                        row="species",
                        col="sex"
                            )

# Aplicar o histograma em cada gráfico - Aplica um histograma em cada gráfico, usando a variável bill_length_mm
grid_species.map(
    sns.histplot,
    "bill_length_mm"
)

# Ajustar títulos
grid_species.set_titles("{row_name} - {col_name}")

# Ajustar nomes dos eixos
grid_species.set_axis_labels(
    "Comprimento do bico (mm)",
    "Quantidade de pinguins"
)

grid_species.axes[0, 0].annotate(
    "Exemplo de anotação",
    xy=(40, 20),
    xytext=(45, 22),
    arrowprops=dict(facecolor="black", shrink=0.05)
)

grid_species.axes[0, 1].annotate(
    "Exemplo de anotação",
    xy=(40, 20),
    xytext=(45, 15),
    bbox=dict(boxstyle="round", facecolor="black", alpha=0.5),
)

plt.show()
```
